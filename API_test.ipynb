```python
#import statements
import streamlit as st
from PIL import Image
import requests
from io import BytesIO
st.set_page_config(page_title="Mars Rover Images", page_icon=":computer:")

# Title & Intro
st.title("Hello, User!")
st.write("""This website will display you the latest images from the curiosity rover
""")


# this code calls the NASA picture API
api_start= "https://api.nasa.gov/mars-photos/api/v1/rovers/curiosity/photos?earth_date="
api_date = "2024-2-19"
api_date_rest= "&api_key="
api_key = "DEMO_KEY"
api_url= api_start+api_date+api_date_rest+api_key
response = requests.get(api_url)
complete_json =response.json()
picture_urls = []
for i in range(len(complete_json["photos"])):
    picture_urls.append(complete_json["photos"][i]['img_src'])

    



def display_image_from_url(url):
    response = requests.get(url)
    image = Image.open(BytesIO(response.content))
    st.image(image, caption='Image from URL', use_column_width=True)

# Streamlit UI
st.title('NASA Curiosity pictures MARS pictures')


# Display image if URL is provided
col1, col2 = st.columns(2)
for i, url in enumerate(picture_urls):
    if i % 2 == 0:
        with col1:
            display_image_from_url(url)
    else:
        with col2:
            display_image_from_url(url)
```
